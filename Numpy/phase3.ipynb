> ## **Advance operation with Bussiness Examples**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("======== Zomato Sales Analysis ========")
print("\n Sales date shape",sales_data.shape)
print("\n Sample data for 1st 3 restro: \n",sales_data[:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:,1:],axis=0)
print(yearly_total)

# mininum sales per restro 
min_sale = np.min(sales_data[:,1:],axis=1) # row
print(min_sale)

# maximum sales per year
max_sale = np.max(sales_data[:,1:],axis=0) # col
print(max_sale)

# Average sales per restro
avg_sale = np.mean(sales_data[:,1:],axis=1)
print(avg_sale)
```

```python
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)

plt.figure(figsize=(8,6)) # set a size of graph
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restro")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("vector addition : ", vector1+vector2)
print("Vector multiplication : ", vector1 * vector2)
print("Dot product : ", np.dot(vector1,vector2))
```

```python
angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

```python
restro_type = np.array(['biryani','chinese','pizza','burger','cafe'])
vectorized = np.vectorize(str.upper)
print(vectorized(restro_type))
```
