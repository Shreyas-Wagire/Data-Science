```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ipl-dataset.csv')
print(data)
```

```python
top_batsman = data[data['Runs']==data['Runs'].max()]
print(top_batsman['Runs'])
```

```python
top_bowler = data[data['Wickets']==data['Wickets'].max()]
print(top_bowler['Player'])
```

```python
group_team = data.groupby('Team').agg({'Runs':'sum','Wickets':'sum'}).reset_index()
print(group_team)
```

```python
team_performance = data.groupby('Team').agg({'Runs': 'sum', 'Wickets': 'sum'}).reset_index()

team_performance.plot(kind='bar', x='Team', y=['Runs', 'Wickets'], figsize=(10, 5))
plt.title('Team Performance in IPL')
plt.xlabel('Team')
plt.ylabel('Total Runs and Wickets')
plt.show()
```
