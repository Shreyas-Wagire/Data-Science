```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('weather.csv')
print(file)
```

```python
Cold_day = file[file['Temperature']==file['Temperature'].min()]
print(Cold_day['Date'].values[0])
```

```python
host_day = file[file['Temperature']== file['Temperature'].max()]
print(host_day['Date'].values[0])
```

```python
file["Temp_MA5"] = file["Temperature"].rolling(window=5).mean()
print(file)
```

```python
plt.figure(figsize=(10,6))
plt.plot(file["Date"], file["Rainfall(mm)"], marker='o', linestyle='-', linewidth=2)

plt.title("Rainfall Trend (Daily)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Rainfall (mm)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
