```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Country": ["India", "USA", "Brazil", "UK", "Russia", "France", "Germany", "Italy", "Spain", "Canada"],
    "Total_Cases": [45000000, 95000000, 35000000, 24000000, 18000000, 22000000, 15000000, 13000000, 12000000, 9000000],
    "Total_Recoveries": [44000000, 92000000, 33000000, 23000000, 17000000, 21000000, 14500000, 12500000, 11800000, 8800000],
    "Total_Deaths": [530000, 1040000, 680000, 170000, 360000, 150000, 120000, 130000, 105000, 26000]
}

covid_df = pd.DataFrame(data)
print(covid_df)
```

```python
growth_rate = (covid_df['Total_Cases']-covid_df['Total_Cases'].shift(1))/ covid_df['Total_Cases'].shift(1) * 100
print(growth_rate)
```

```python
# RecoveryRate=CasesRecoveries​×100
recovery_rate = (covid_df['Total_Recoveries'] / covid_df['Total_Cases']) * 100
print(recovery_rate.head(5))
```

```python
normalized_count = covid_df['Total_Cases'] / covid_df['Total_Cases'].max()
print(normalized_count)
```

```python
# Line graph → growth trend over time.
# Bar graph → compare recovery rates.

plt.figure(figsize=(10, 5))
plt.plot(covid_df['Country'], growth_rate, marker='o', label='Growth Rate (%)')
plt.title('COVID-19 Growth Rate by Country')
plt.xlabel('Country')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
